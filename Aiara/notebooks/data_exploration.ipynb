```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
```

```python
# Read data
df = pd.read_csv('data/med_recon.csv')
```

```python
df.head()
```

```python
# Visual search for correlations
cols_select = ['kidney_function', 'hospitalization', 'n_ae_visits', 'specialisms_visits', 
               'changed_police_based_reconciliation']
g = sns.PairGrid(df, vars=cols_select,
                 hue='changed_police_based_reconciliation')
g.map(sns.scatterplot)
plt.show()
```
